# Tasks 4: Customer Segmentation Visualization

### 1) Visualize customer distribution by tenure using pie or donut charts. (e.g., 0-12 months, 13-36 months, 37+ months).

```python
# Define tenure categories
bins = [0, 12, 36, df['tenure'].max()]
labels = ['0-12 months', '13-36 months', '37+ months']

# Create tenure category column
df['tenure_category'] = pd.cut(df['tenure'], bins=bins, labels=labels, right=True)

# Count customers in each category
tenure_counts = df['tenure_category'].value_counts()
```

```python
# Plot donut chart
plt.figure(figsize=(8, 8))
colors = ['#ff9999','#66b3ff','#99ff99']
plt.pie(tenure_counts, labels=tenure_counts.index, autopct='%1.1f%%', colors=colors, wedgeprops=dict(width=0.6))
plt.title("Customer Distribution by Tenure")
plt.show()
```

### 2) Use a clustered bar chart to compare average monthly charges across tenure categories,


```python
# Create tenure category column
df['tenure_category'] = pd.cut(df['tenure'], bins=bins, labels=labels, right=True)

# Compute average monthly charges per tenure category
avg_monthly_charges = df.groupby('tenure_category')['monthlycharges'].mean().reset_index()

# Plot clustered bar chart
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='tenure_category', y='monthlycharges', data=avg_monthly_charges, palette='viridis')

# Customize plot
plt.title("Average Monthly Charges by Tenure Category")
plt.xlabel("Tenure Category")
plt.ylabel("Average Monthly Charges")

# Annotate the bars with the exact numbers
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=12, color='black', 
                xytext=(0, 5), textcoords='offset points')

# Find highest and lowest bars
max_bar = avg_monthly_charges['monthlycharges'].idxmax()
min_bar = avg_monthly_charges['monthlycharges'].idxmin()

# Highlight the highest bar
ax.patches[max_bar].set_edgecolor('red')
ax.patches[max_bar].set_linewidth(2)
ax.annotate(f'Highest: {avg_monthly_charges["monthlycharges"][max_bar]:.2f}', 
            (ax.patches[max_bar].get_x() + ax.patches[max_bar].get_width() / 2., 
             ax.patches[max_bar].get_height() + 2), 
            ha='center', va='bottom', fontsize=12, color='red')

# Highlight the lowest bar
ax.patches[min_bar].set_edgecolor('blue')
ax.patches[min_bar].set_linewidth(2)
ax.annotate(f'Lowest: {avg_monthly_charges["monthlycharges"][min_bar]:.2f}', 
            (ax.patches[min_bar].get_x() + ax.patches[min_bar].get_width() / 2., 
             ax.patches[min_bar].get_height() + 2), 
            ha='center', va='bottom', fontsize=12, color='blue')

plt.show()
```

# Tasks 5: Advanced Analysis

```python
# Load dataset
df = pd.read_csv("C:\\Users\\dell\\Pictures\\Shrunk\\Virtula Intership\\Telco_Customer_Churn_Cleaned.csv")

# Convert 'totalcharges' to numeric (handle missing values if any)
df["totalcharges"] = pd.to_numeric(df["totalcharges"], errors='coerce')
df.fillna({"totalcharges": df["totalcharges"].median()}, inplace=True)

# Convert 'churn' to binary
df["churn"] = df["churn"].map({"Yes": 1, "No": 0})

# Create tenure bins
tenure_bins = [0, 12, 24, 36, 48, 60, df["tenure"].max()]
tenure_labels = ["0-12", "13-24", "25-36", "37-48", "49-60", "60+"]
df["TenureGroup"] = pd.cut(df["tenure"], bins=tenure_bins, labels=tenure_labels, right=False)

# Group by tenure and compute statistics
churn_by_tenure = df.groupby("TenureGroup").agg({
    "churn": "mean", 
    "monthlycharges": ["mean", "median", "std"], 
    "totalcharges": ["mean", "std"], 
    "customerid": "count"
}).reset_index()
churn_by_tenure.columns = ["TenureGroup", "churnRate", "monthlychargesMean", "monthlychargesMedian", "monthlychargesStd", "totalchargesMean", "totalchargesStd", "CustomerCount"]

# Plot churn rate by tenure group
plt.figure(figsize=(10,5))
sns.barplot(x="TenureGroup", y="churnRate", hue="TenureGroup", data=churn_by_tenure, legend=False)
plt.title("Churn Rate by Tenure Group")
plt.ylabel("Churn Rate")
plt.show(block=True)
```

```python
# Group by tenure and contract type
tenure_contract_analysis = df.groupby(["TenureGroup", "contract"]).agg({"churn": "mean", "customerid": "count"}).reset_index()
tenure_contract_analysis.columns = ["TenureGroup", "ContractType", "churnRate", "CustomerCount"]

plt.figure(figsize=(12,6))
sns.barplot(x="TenureGroup", y="churnRate", hue="ContractType", data=tenure_contract_analysis)
plt.title("Churn Rate by Tenure and Contract Type")
plt.ylabel("Churn Rate")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[["tenure", "monthlycharges", "totalcharges", "churn"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Churn rate by demographics
demo_cols = ["gender", "seniorcitizen", "partner", "dependents"]
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()
for i, col in enumerate(demo_cols):
    sns.barplot(x=col, y="churn", hue=col, data=df, ax=axes[i], legend=False)
    axes[i].set_title(f"Churn Rate by {col}")
plt.tight_layout()
plt.show()
```

```python
# Stacked bar chart for churn vs. non-churn by demographics
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()  # Flatten to 1D array

for i, col in enumerate(demo_cols):
    churn_dist = df.groupby([col, "churn"]).size().unstack()
    churn_dist.plot(kind="bar", stacked=True, ax=axes[i])
    axes[i].set_title(f"Churn Distribution by {col}")
plt.tight_layout()
plt.show()
```

```python
# Pie charts for categorical variables
fig, axes = plt.subplots(1, 3, figsize=(15,5))
for ax, col in zip(axes, ["contract", "paymentmethod", "churn"]):
    df[col].value_counts().plot.pie(autopct="%1.1f%%", ax=ax, startangle=90)
    ax.set_title(f"Distribution of {col}")
    ax.set_ylabel("")
plt.show()
```

```python
# Churn by contract type
plt.figure(figsize=(8,5))
sns.barplot(x="contract", y="churn", hue="contract", data=df, legend=False)
plt.title("Churn Rate by Contract Type")
plt.show()
```

```python
# Churn by payment method
plt.figure(figsize=(10,5))
sns.barplot(x="paymentmethod", y="churn", hue="paymentmethod", data=df, legend=False)
plt.xticks(rotation=45)
plt.title("Churn Rate by Payment Method")
plt.show()
```

```python
# Churn trend over tenure
plt.figure(figsize=(12,6))
sns.lineplot(x="tenure", y="churn", data=df, estimator="mean")
plt.title("Churn Trend Over Time")
plt.xlabel("Tenure (Months)")
plt.ylabel("Churn Rate")
plt.show()
```

### Key Findings from Telecom Churn Analysis


Churn Rate by Tenure:
Shorter tenure customers (0-12 months) have the highest churn rate.
Longer tenure customers (60+ months) are more likely to remain.

Churn by Contract Type:
Month-to-month contracts experience the highest churn.
Two-year contracts exhibit much lower churn, suggesting contract duration affects retention.

Churn by Payment Method:
Customers paying by electronic check have a higher churn rate.
Automatic payments (credit card/bank transfer) result in lower churn.

Demographic Insights:
Older customers churn more than younger ones.
Dependent-free or partner-free customers churn moreâ€”presumably because they have fewer long-term commitments.

Financial Factors:
Higher monthly fees are associated with higher churn.
But overall fees don't strongly forecast churnâ€”implying churn is motivated by repeat costs rather than cumulative spending.

Overall Pattern:
Higher-cost customers with flexible contracts churn more quickly.
Promoting longer contracts and discounts for automatic payments may be able to cut churn.
